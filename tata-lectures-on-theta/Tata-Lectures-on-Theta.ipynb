# Introduction and motivation: theta functions in one variable

## 1. Definition of $\vartheta(z, \tau)$ and its periodicity in $z$.

The central character in our story is the analytic function $\vartheta(z, \tau)$ in
2 variables defined by

$$
\vartheta(z, \tau) = \sum_{n \in \mathbb{Z}} \exp{(\pi i n^2 \tau + 2 \pi inz)}
$$
where $z \in \mathbb{C}$ and $\tau \in H$, the upper half plane $\operatorname{Im} \tau > 0$.
It is immediate that the series converges absolutely and uniformly on compact sets; in fact, if
$$
\left|\operatorname{Im} z \right|<c \text{ and } \operatorname{Im} T > \epsilon
$$
then
$$
\left|\exp{(\pi in^2 \tau + 2\pi inz)} \right|< (\exp{ - \pi \epsilon})^{n^2} \cdot (\exp{2\pi c})^n
$$
hence, if $n_0$ is chosen so that
$$
(\exp{-\pi \epsilon})^{n_0} \cdot (\exp{ 2\pi c}) < 1,
$$
then the inequality
$$
\left|\exp{(\pi in^2 \tau + 2\pi inz)} \right| < (\exp{ - \pi \epsilon})^{n(n-n_0)}
$$
shows that the series converges and that too very rapidly.\
We may think of this series as the Fourier series for a function in $z$,
periodic with respect to $z \mapsto z+1$,
$$
\vartheta(z, \tau) =  \sum_{n \in \mathbb{Z}} a_n(\tau)\exp{(2 \pi inz)}, \; a_n(\tau) = \exp{(\pi in^2 \tau)}
$$
which displays the obvious fact that
$$
\vartheta(z+1, \tau) = \vartheta(z, \tau)
$$

(Page 2)\
The peculiar form of the Fourier coefficients is explained by
examining the periodic behaviour of $\vartheta$ with respect to $z \mapsto z + \tau$ : \
thus we have
$$
\begin{align}
\vartheta(z+\tau,\tau)
&= \sum_{n \in \mathbb{Z}} \exp{(\pi i n^2 \tau + 2 \pi in(z+\tau))} \\
&= \sum_{n \in \mathbb{Z}} \exp{(\pi i (n+1)^2 \tau -\pi i \tau + 2 \pi inz)} \\
&=  \sum_{m \in \mathbb{Z}} \exp{(\pi i m^2 \tau -\pi i \tau + 2 \pi imz - 2 \pi iz)} \\
&= \exp{(-\pi i\tau - 2\pi iz)}\cdot \vartheta(z, \tau)
\end{align}
$$
so that $\vartheta$ has a kind of periodic behaviour with respect to the lattice
$\Lambda_{\tau} \subset \mathbb{C}$ generated by $1$ and $\tau$ . In fact the 2 periodicities together are
easily seen to give:
$$
\vartheta(z +a\tau+b,\tau) = \exp{(-\pi ia^2 \tau - 2\pi iaz)}\vartheta(z, \tau).
$$
 Conversely, suppose that we are looking for entire functions $f(z)$
with the simplest possible quasi-periodic behaviour w. r. t. $\Lambda_{\tau}$ : we know
by Liouville's theorem that $f$ cannot actually be periodic in $1$ and $\tau$, so
we may try the simplest more general possibilities:
$$
f(z+1) = f(z) \text{ and } f(z +\tau) = \exp{(az+b)}\cdot f(z) .
$$
By the first, we expand $f$ in a Fourier series
$$
f(z) = \sum_{n \in \mathbb{Z}}  a_n \exp{(2\pi inz)}, \; a_n \in \mathbb{C}.
$$
Writing $f(z+\tau+1)$ in terms of $f(z)$ by combining the functional equations
in either order, we find that

(Page 3)\
$$
f(z+\tau +1) = f(z +\tau) = \exp{(az+b)} \cdot f(z)
$$
and also
$$
\begin{align}
f(z +\tau +1) &= \exp{(a(z+1) +b)} f(z+1)\\
 &= \exp{a} \cdot \exp{(az + b)} f (z)
\end{align}
$$
hence $a = 2\pi ik$ for some $k \in \mathbb{Z}$. Now substituting the Foruier series
into the second equation, we find that
$$
\begin{align}
\sum_{n \in \mathbb{Z}}  a_n \exp{(2\pi inz)}
&= f(z +\tau) \\
&= \exp{(2\pi ikz + b)} f(z) \\
&= \sum_{n \in \mathbb{Z}} a_n \exp{(2\pi i(n+ k)z)} \cdot \exp{b} \\
&= \sum_{n \in \mathbb{Z}} a_{n-k}  \exp{b} \cdot \exp{(2\pi inz)} 
\end{align}
$$
Or, equivalently, for all $n \in \mathbb{Z}$, we have
$$
(*)\quad a_n = a_{n-k} \exp{(b-2\pi in \tau)} .
$$
If $k = 0$, this shows immediately that $a_n \ne 0$ for __at most one__ $n$ and we
have the uninteresting possibility that $f(z) = \exp{(2\pi iz)}$. If $k \ne 0$, we get
a recursive relation for solving for $a_{n+kp}$ in terms of $a_n$ for all $p$. For
instance, if $k = -1$ , we find easily that
$$
a_n = a_0 \exp{ (-nb +\pi in(n-1)\tau)} \text{ for all } n\in \mathbb{Z} .
$$
This means that
$$
\begin{align}
f(z) &= a_0 \sum_{n \in \mathbb{Z}} \exp{ (-nb-\pi in \tau)}  \exp{(\pi in^2 \tau + 2\pi inz)} \\
&=  a_0 \; \vartheta\left(- z - \frac{1}{2}\tau -\frac{b}{2}\pi i , \tau\right)
\end{align}
$$

(Page 4) \
If $k >0$, the recursion relation ($*$) leads to rapidly growing coefficients
$a_n$ and hence there ar e no such entire functions $f(z)$. On the other hand,
if $k < -1$ , we will find a $|k |$ -dimensional vector space of possibilities for
$f(z)$ that will be studied in detail below. This explains the significance of
$\vartheta(z, \tau)$ as an entire function of $z$ for fixed $\tau$, i.e. , $\vartheta(z, \tau)$ is the most
general entire function with 2 quasi-periods.
