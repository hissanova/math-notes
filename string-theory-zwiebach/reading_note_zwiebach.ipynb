## §8.4 The Complete Momentum Current

(8.39) is very intriguing. The currents live on the worldsheet and only on it: their indices are world-sheet indices, and their arguments are world-sheet coordinates. As seen in spacetime, the currents vanish everywhere except on the surface traced out by the string.

The integral in (8.36) is then an integral over the strings, the lines of constant time as seen by the chosen Lorentz observer.

In the static gauge $t = \tau$, the conservation becomes,
$$
\frac{d p_{\mu}}{dt} = 0
$$
The Lorentz observer confirms that momentum is conserved in time.

Equation (8.39), together with (8.36), tells us that for any fixed world-sheet parameteriza-
tion, we can compute a unique quantity pμ using any line of constant τ . This quantity must
coincide with the time-independent momentum pμ obtained using the static gauge, as we
now explain.

We can show __any curve__ on the world-sheet can be used to calculate the conserved momentum $p_{\mu}$

We now show how to generalize (8.36) to curves
- open string: the curve must stretch from one boundary to the other.
- closed string: the curve must be a closed non-contractible curve.






We can use Green's theorem
$$
\oint_{\partial R} P d\tau - Q d\sigma = \int_{R} \left( \frac{\partial P}{\partial \tau} + \frac{\partial Q}{\partial \sigma} \right) d\tau d\sigma
$$
on the vector field $\left(\mathcal{P}^{\tau}_{\mu}, \mathcal{P}^{\sigma}_{\mu}\right)$, where $R$ is a simply connected region, $P,Q$ smooth functions.

Setting
$$
p_{\mu}(\partial R) := \oint_{\partial R} \mathcal{P}^{\tau}_{\mu} d\sigma - \mathcal{P}^{\sigma}_{\mu} d\tau,
$$
we can immediately see that this is equal to zero.

Now, we need to consider two separate cases:

(1) Open string:\
Consider $\partial R = \tilde{\gamma} - \beta - \gamma + \alpha$ as in the right diagram of the fig. 8.1.
Since $\alpha$ and $\beta$ are curves where $\sigma$ and $\mathcal{P}^{\sigma}_{\mu}$ both vanish, so $\int_{\alpha \text{ or } \beta} (-) = 0$.
Hence, 
$$
\int_{\gamma} \mathcal{P}^{\tau}_{\mu} d\sigma - \mathcal{P}^{\sigma}_{\mu} d\tau
= \int_{\tilde{\gamma}} \mathcal{P}^{\tau}_{\mu} d\sigma - \mathcal{P}^{\sigma}_{\mu} d\tau.
$$
This proves $p_{\mu}(\gamma)$ for any curve $\gamma$ which connects the $\sigma=0$ and $\sigma=\sigma_1$ boundaries of the world-sheet has the same value.

(2) Closed string:\
This is the case where we identify $\alpha$ and $\beta$ whence $\left(\int_{\alpha} - \int_{\beta}\right)\left(-\right) = 0$.

# §8.5 Lorentz Symmetry and Associated Currents
By construction, the action of the relativistic string is Lorentz invariant since it is written in the terms which are Lorentz invariant namely .

$$
(A, B) = \eta_{\mu\nu} A^{\mu}B^{\nu}
$$

The Lorentz transformations are linear transformations of the coordinates $X^{\mu}$ that leave the quadratic form $\eta_{\mu\nu}X^{\mu}X^{\nu}$ invariant. 

Every infinitesimal linear transformation is of the form $X^{\mu} \rightarrow X^{\mu} + \delta X^{\mu}$, where

$$
\delta X^{\mu} = \epsilon^{\mu\nu}X_{\nu}.
$$
Here $\epsilon^{\mu\nu}$ is an infinitesimal matrix. If we require $\epsilon^{\mu\nu} =-\epsilon^{\nu\mu}$ then $\delta(\eta_{\mu\nu}X^{\mu}X^{\nu}) = 0$. 

We can show that this infinitesimal transformation preserves the string action by showing all the terms of the form
$$
\eta_{\mu\nu} \frac{\partial X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial X^{\nu}}{\partial \xi^{\beta}}
$$
remain zero, since the action consists only of these terms.

$
\begin{align}
\delta\left(\eta_{\mu\nu} \frac{\partial X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial X^{\nu}}{\partial \xi^{\beta}}\right) 
&= \eta_{\mu\nu} \left( \frac{\partial \delta X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial X^{\nu}}{\partial \xi^{\beta}} +
                        \frac{\partial X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial \delta X^{\nu}}{\partial \xi^{\beta}} \right) \\
&= \eta_{\mu\nu} \left( \epsilon^{\mu\rho}\frac{\partial X_{\rho}}{\partial \xi^{\alpha}} \frac{\partial X^{\nu}}{\partial \xi^{\beta}} +
                        \epsilon^{\nu\rho}\frac{\partial X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial X_{\rho}}{\partial \xi^{\beta}} \right) \\
&=  \epsilon_{\mu\rho}\frac{\partial X^{\rho}}{\partial \xi^{\alpha}} \frac{\partial X^{\mu}}{\partial \xi^{\beta}} +
    \epsilon_{\nu\rho}\frac{\partial X^{\nu}}{\partial \xi^{\alpha}} \frac{\partial X^{\rho}}{\partial \xi^{\beta}}  \\
&=  (\epsilon_{\mu\rho} + \epsilon_{\rho\mu}) \frac{\partial X^{\mu}}{\partial \xi^{\alpha}} \frac{\partial X^{\rho}}{\partial \xi^{\beta}}  \\
&=0
\end{align}
$

$$
\begin{align}
\epsilon^{\mu\nu} j^{\alpha}_{\mu\nu} 
&= \frac{\partial \mathcal{L}}{\partial(\partial_{\alpha}X^{\mu})}\delta X^{\mu}\\
&= \mathcal{P}^{\alpha}_{\mu}\epsilon^{\mu\nu}X_{\nu} \\
&= -\frac{1}{2}\epsilon^{\mu\nu} (X_{\mu}\mathcal{P}^{\alpha}_{\nu} - X_{\nu} \mathcal{P}^{\alpha}_{\mu}) \\
\end{align}
$$
We define the currents 
$$
\mathcal{M}^{\alpha}_{\mu\nu} = X_{\mu}\mathcal{P}^{\alpha}_{\nu} - X_{\nu} \mathcal{P}^{\alpha}_{\mu}.
$$
By construction,
$$ \mathcal{M}^{\alpha}_{\mu\nu} = - \mathcal{M}^{\alpha}_{\nu\mu}.$$
it satisfies 
$$ \frac{\partial \mathcal{M}^{\tau}_{\mu\nu}}{\partial \tau} + \frac{\partial \mathcal{M}^{\sigma}_{\mu\nu}}{\partial \sigma} = 0.$$
The charge is given by 
$$
M_{\mu\nu} = \int_{\gamma} \mathcal{M}^{\tau}_{\mu\nu} d\sigma - \mathcal{M}^{\sigma}_{\mu\nu} d\tau.
$$

Note that $M_{\mu\nu}$ is contour independent. \
For the closed string case is similar to the momentum charge, and for the open string case, we notte that $\mathcal{M}^{\sigma}_{\mu\nu}$ contains only the terms of $\mathcal{P}^{\sigma}$ and ${P}^{\sigma}$ vanishes on the world-sheet boundaries.

 $M_{\mu\nu}$ has six charges:
- $M_{0i}$: for three boosts
- $M_{ij}$: for three spatial rotations

The charge associated with the boosts are:
$$
M^{0i} = \int d\sigma \left( ct \mathcal{P}^{\tau i}  - X^{i} \mathcal{P}^{\tau 0}\right) = ct p^i - \int d\sigma X^i \mathcal{P}^{\tau 0}.
$$
Multiplying by $\frac{c}{E}$, ($E$ is the conserved energy of the string)
$$
\frac{cM^{0i}}{E} = t\frac{c^2 p^i}{E} - \frac{1}{E} \int d\sigma X^i c\mathcal{P}^{\tau 0}.
$$
Since $c\mathcal{P}^{\tau 0}$ is the energy density along the string, the last term on the right-hand side can be identified with the time-dependent position $X^i_{cm}(t)$ of the centre of mass (energy)of the string. We thus have,
$$
X^i_{cm}(t) = - \frac{cM^{0i}}{E} + t\frac{c^2 p^i}{E}.
$$
$\frac{c^2 p^i}{E}$ has the interpretation of the centre of mass velocity.

## 8.6 The Slope Parameter $\alpha^{\prime}$

A parameter called slope parameter has an historical importance and most of the modern string theory is written using this parameter instead of the tension $T_0$.

It is defined as 
$$
\frac{J}{\hbar} = \alpha^{\prime} E^2.
$$
So 
$$
[\alpha^{\prime}] = \frac{1}{[E^2]}
$$
It can be shown, considering a straight open string of energy E which rotates rigidly on the (x, y) plane (details are on the textbook)
$$
J = \frac{1}{2\pi T_0 c} E^2.
$$
Hence, the two parameters are related as
$$
\alpha^{\prime} = \frac{1}{2\pi T_0 \hbar c}
$$
