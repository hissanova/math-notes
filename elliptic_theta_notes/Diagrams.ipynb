# Ellipse
Canonical form:
$$ \frac{x^{2}}{a^{2}} +\frac{y^{2}}{b^{2}}=1 \quad with \; a \ge b.$$
parametric form:
$$\mathbf{p} =\left(
\begin{array}{c}
 x \\
 y
\end{array} \right)
=\left(
\begin{array}{c}
 a \sin{\varphi} \\
 b \cos{\varphi}
\end{array} \right)
\;, where \; 0 \le \varphi < 2 \pi
$$

```python
var('t')

color = 'black'

a = 1.5
b = 1
angle = 0.5 

g = Graphics()

g += parametric_plot([a * sin(t), b * cos(t)], [t, 0, 2 * pi], color=color)
g += parametric_plot([a * sin(t), b * cos(t)], [t, 0, angle ], color=color, thickness=3)
g += point([a * sin(angle), b * cos(angle)], size=40, color=color)
g += line([[0, 0], [a * sin(angle), b * cos(angle)]], color=color)

text_pos = 0.1
g += text('$p$', [a * sin(angle) + text_pos , b * cos(angle) + text_pos], color=color, fontsize='x-large')
g += text('$s$', [a * sin(angle/2.5) + text_pos , b * cos(angle/2.5) + text_pos], color=color, fontsize='x-large')
g.axes_labels(['$x$', '$y$'])
g.matplotlib(ticks=[[],[]])
g.show(ticks=[[],[]], figsize=[5,5])
g.save("ellipse_generic.png", ticks=[[],[]], figsize=[5,5])
```
