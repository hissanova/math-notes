# Ellipse
Canonical form:
$$ \frac{x^{2}}{a^{2}} +\frac{y^{2}}{b^{2}}=1 \quad with \; a \ge b.$$
parametric form:
$$\mathbf{p} =\left(
\begin{array}{c}
 x \\
 y
\end{array} \right)
=\left(
\begin{array}{c}
 a \sin{\varphi} \\
 b \cos{\varphi}
\end{array} \right)
\;, where \; 0 \le \varphi < 2 \pi
$$

```python
var('t')

color = 'black'

a = 1.5
b = 1
angle = 0.5 

g = Graphics()

g += parametric_plot([a * sin(t), b * cos(t)], [t, 0, 2 * pi], color=color)
g += parametric_plot([a * sin(t), b * cos(t)], [t, 0, angle ], color=color, thickness=3)
g += point([a * sin(angle), b * cos(angle)], size=40, color=color)
g += line([[0, 0], [a * sin(angle), b * cos(angle)]], color=color)

text_pos = 0.1
g += text('$\\mathbf{p}$', [a * sin(angle) + text_pos , b * cos(angle) + text_pos], color=color, fontsize='x-large')
g += text('$s$', [a * sin(angle/2.5) + text_pos , b * cos(angle/2.5) + text_pos], color=color, fontsize='x-large')
g.axes_labels(['$x$', '$y$'])
g.matplotlib(ticks=[[],[]])
g.show(ticks=[[],[]], figsize=[5,5])
g.save("ellipse_generic.png", ticks=[[],[]], figsize=[5,5])
```

# Curves of Cassini
A curve of Cassini is defined as a collection of points $P(x,y)$ whose product of the distances from two given points $F_1, F_2$ , called foci, is a constant $a^2$. For simplicity, let us suppose $F_1 = (p,0), F_2 = (-p, 0)$ for some $p>0$ and let $r_1 = \overline{PF_1}, r_2 = \overline{PF_2}$, then
$$ r_1 r_2 = a^2. $$

```python
var('x')

color = 'black'

x_0 = 1
y_0 = 1

a = 1
p = 2.1
text_pos = 0.05

g = Graphics()

g += point([x_0, y_0], size=40, color=color)
g += point([p, 0], size=40, color=color)
g += point([-p, 0], size=40, color=color)
text_pos = 0.2
g += text('$P(x,y)$', [x_0 + text_pos , y_0 + text_pos], color=color, fontsize='x-large')
g += text('$p$', [p, 0 - text_pos], color=color, fontsize='x-large')
g += text('$-p$', [-p, 0 - text_pos], color=color, fontsize='x-large')
g += line([[p, 0], [x_0,y_0]], color=color)
g += line([[-p, 0], [x_0,y_0]], color=color)

g += text('$r_1$', [(p + x_0)/2 + text_pos , (0 + y_0)/2 + text_pos], color=color, fontsize='x-large')
g += text('$r_2$', [(-p + x_0)/2 - text_pos , (0 + y_0)/2 + text_pos], color=color, fontsize='x-large')

g.axes_labels(['$x$', '$y$'])
g.axes_range(ymin=-0.3)
g.matplotlib(ticks=[[],[]])
g.show(ticks=[[],[]], figsize=[5,5], aspect_ratio=1)
g.save("cassini_generic_point.png", ticks=[[],[]], figsize=[5,5], aspect_ratio=1)
```
